```python
import pandas as pd
```

```python
df = pd.read_parquet('gg21_donors_scored.parquet')
df
```

```python
df = df[['Address', 'aggregate_score']]
df.columns = ['address', 'rawScore']
df
```

```python
df_22 = pd.read_csv('gg22_donors_scored.csv')
df_22 = df_22[['Address', 'aggregate_score']]
df_22.columns = ['address', 'rawScore']
df_22
```

```python
df_22 = df_22[['Address', 'aggregate_score']]
df_22.columns = ['address', 'rawScore']
df_22
```

```python
print("df stats:")
print(f"- Total rows: {len(df)}")
print(f"- NaN counts: {df['rawScore'].isna().sum()}")
print(f"- NaN percentage: {(df['rawScore'].isna().sum() / len(df) * 100):.2f}%")

print("\ndf_22 stats:")
print(f"- Total rows: {len(df_22)}")
print(f"- NaN counts: {df_22['rawScore'].isna().sum()}")
print(f"- NaN percentage: {(df_22['rawScore'].isna().sum() / len(df_22) * 100):.2f}%")
```

```python
df_22 = df_22.dropna(subset=['rawScore'])

print("\ndf_22 stats:")
print(f"- Total rows: {len(df_22)}")
print(f"- NaN counts: {df_22['rawScore'].isna().sum()}")
print(f"- NaN percentage: {(df_22['rawScore'].isna().sum() / len(df_22) * 100):.2f}%")
```

```python
df = pd.concat([df, df_22], ignore_index=True)
print("df stats:")
print(f"- Total rows: {len(df)}")
print(f"- NaN counts: {df['rawScore'].isna().sum()}")
print(f"- NaN percentage: {(df['rawScore'].isna().sum() / len(df) * 100):.2f}%")
```

```python
# Convert addresses to lowercase
df['address'] = df['address'].str.lower()

# Check for duplicate addresses initially
duplicate_count = df['address'].duplicated().sum()
print(f"\nDuplicate addresses before deduping: {duplicate_count}")

if duplicate_count > 0:
    print("\nExample duplicates before deduping:")
    duplicates = df[df['address'].duplicated(keep=False)].sort_values('address')
    print(duplicates.head(10))
    
    # Drop duplicates keeping df_22 entries
    df = df.drop_duplicates(subset='address', keep='last')
    
    # Check for duplicates again
    duplicate_count = df['address'].duplicated().sum()
    print(f"\nDuplicate addresses after deduping: {duplicate_count}")
    
    if duplicate_count > 0:
        print("\nExample duplicates after deduping:")
        duplicates = df[df['address'].duplicated(keep=False)].sort_values('address')
        print(duplicates.head(10))
```
